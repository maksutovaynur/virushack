```python
input_text = S.keywords(S.lemm_h('Голова болит очень сильно'))
display(input_text)
Sd2['tmp'] = Sd2['Tf-Idf'].apply(lambda x: S.multiply2stats(input_text, x, max_dist=0))
Sd2['tmp_sum'] = Sd2['tmp'].apply(S.sum_stat)
sSd2 = Sd2.sort_values('tmp_sum', ascending=False)
candDeseases = sSd2.head(3)
# candDeseases[['Name', 'Targ', 'tmp_sum']].head(1)
candDeseasesInfo = dict(candDeseases.loc[:, 'Name'].apply(lambda x: S.keywords(T.lemm(x))).agg(T.sum_dicts))
display(candDeseasesInfo)
```

```python
Ph2['tmp'] = Ph2['Tf-Idf'].apply(lambda x: S.multiply2stats(candDeseasesInfo, x, max_dist=0))
Ph2['tmp_sum'] = Ph2['tmp'].apply(S.sum_stat)
Ph2.sort_values('Subst', inplace=True)
sPh2 = Ph2.groupby('Subst').agg({'Name': list, 'Targ': 'first', 'tmp': 'first', 'tmp_sum': 'first'}).reset_index()
sPh2 = sPh2.sort_values('tmp_sum', ascending=False)
candPharm = sPh2.head(5)
candPharm = candPharm.loc[candPharm.loc[:, 'tmp_sum'] > 0]
# candPharmInfo = dict(candPharm.loc[:, 'Tf-Idf'].agg(T.sum_dicts))
# display(candPharmInfo)
display(candPharm)
display(Ph2.head(30))
```
