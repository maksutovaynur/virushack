```python
import sys
sys.path.append('..')
sys.path.append('../lib')
import stats as S
import json
from importlib import reload
import _pickle as p
import pandas as pd
from IPython.display import display, HTML
S = reload(S)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
display(HTML("<style>.container{width: 100%}</style>"))
```

```python
with open('word_importance.json', 'r') as f:
    word_importance = json.loads(f.read())
with open('db2.dat', 'rb') as f:
    data2 = p.load(f)
THRESHOLD = 1e-1
Ph = pd.DataFrame(data2)
Ph['Lemms'] = Ph['symptom_text'].map(lambda x: S.lemm(x))
Ph['Keywords'] = Ph['Lemms'].map(lambda x: {i: 1 for i, v in S.keywords(x).items()})
Ph['Multiplied'] = Ph['Keywords'].map(lambda x: S.multiply2stats(word_importance, x, max_dist=0))
Ph['Filtered'] = Ph['Multiplied'].map(lambda x: {i: v for i, v in x.items() if v > max(x.values())*THRESHOLD})
Ph['FSum'] = Ph['Filtered'].map(lambda x: sum(x.values()))
Ph['Normalized'] = Ph.apply(lambda x: {i: v/x['FSum'] for i, v in x['Filtered'].items()}, axis=1)
display(Ph.iloc[:2])
    
# Dtr = pd.read_csv('des_symp_translate.csv', sep='\t')
# Dtr['Lemms'] = Dtr['Rus'].map(lambda x: S.lemm(x))
# Dtr['Keywords'] = Dtr['Lemms'].map(lambda x: S.keywords(x))
# display(Dtr.head())

# DmlTest = pd.read_csv("Testing.csv")
# DmlTrain = pd.read_csv("Training.csv")
# display(DmlTest.head())
# display(DmlTrain.head())
```
