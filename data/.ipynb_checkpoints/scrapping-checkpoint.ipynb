```python
def get_soup(url):
    """
    получает распарсенный html объект из юрла
    """
    URL = url
    page = requests.get(URL)
    return BeautifulSoup(page.content, 'html.parser')

def traverse_mkb(url, db):
    """
    Функция которая весь справочник болезней херачит
    """
    sp = get_soup(url)
    for el in sp.find_all('a',  {'class': 'classification__link'}):
        traverse_submkb_rec('http:' + el['href'], db)

def update_db(sp, db):
    """
    Вытаскивает инфу со странички и добавляет в базу
    """
    desease_id = sp.find('h1').text if sp.find('h1') else None
    
    desease_description = sp.find('div', {'class':'news_text'})
    symptoms = []
    if desease_description:
        found_symptoms_header = False
        for p in desease_description.find_all('p'):
            if found_symptoms_header:
                symptoms.append(p.text)
            elif 'симптом' in p.text.lower():
                found_symptoms_header = True
                
    active_shit = []
    for x in sp.find_all('td', {'class':'rest_data'}):
        if x.find('a'):
            href = x.find('a')
            if 'mnn_index_id' in href['href']:
                active_shit.append({'href': 'https:' + href['href'], 'name': href.text})
    
    db.append({'id':desease_id, 'symps':'\n'.join(symptoms), 'actives':active_shit})
            

def traverse_submkb_rec(url, db):
    """
    проходит по страничке, добавляет в базу, проходит рекурсивно по вложенным
    страничкам болезней и добавляет в базу
    """
    sp = get_soup(url)
    update_db(sp, db)
    for el in sp.find_all('a',  {'class': 'subcatlist__link'}):
        traverse_submkb_rec('http:' + el['href'], db)

```

```python
kwds_sorted = sorted(kwds.items(), key=lambda x: -x[1])
kwds_inverted = {i: word_cnt_log_inv/v for i, v in kwds.items()}
v = kwds_inverted.values()
with open('word_importance.json', 'w') as f:
    f.write(json.dumps(kwds_inverted))
min(v), max(v)
```
